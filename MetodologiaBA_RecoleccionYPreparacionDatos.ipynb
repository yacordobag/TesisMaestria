```python
def resumen_raster(path):
    with rasterio.open(path) as src:
        print(f"\n Archivo: {os.path.basename(path)}")
        print("  CRS        :", src.crs)
        print("  Resoluci√≥n :", src.res)
        print("  Tama√±o     :", src.width, "x", src.height)
        print("  N¬∫ bandas  :", src.count)


        banda1 = src.read(1, masked=True)
        vals = banda1.compressed()

        if vals.size > 2000000:
            vals = vals[::100]
        print("  Valor m√≠n. :", float(vals.min()))
        print("  Valor m√°x. :", float(vals.max()))
```

```python
import numpy as np

def rango_valores(path):
    with rasterio.open(path) as src:
        arr = src.read(1).astype(float)
        nodata = src.nodata


        valid = arr[(arr != nodata) & (~np.isnan(arr))]

        print(f"\nArchivo: {os.path.basename(path)}")
        print("Valores v√°lidos:", valid.size)
        print("Min:", valid.min())
        print("Max:", valid.max())

rango_valores(os.path.join(base, "LST_2014_Reproyectado.tif"))
rango_valores(os.path.join(base, "LST_2024_Reproyectado.tif"))
```

```python
import os
import rasterio
from rasterio.warp import reproject, Resampling
import numpy as np

base = "/content/drive/MyDrive/MAESTR√çA CIENCIA DE LOS DATOS Y ANAL√çTICA/TRABAJO DE GRADO/DATOS/DATOS_SATELITALES"

ref_path = os.path.join(base, "LST_2014_Reproyectado.tif")
src_path = os.path.join(base, "LST_2024_Reproyectado.tif")
out_path = os.path.join(base, "LST_2024_match_2014.tif")


with rasterio.open(ref_path) as ref:
    dst_crs = ref.crs
    dst_transform = ref.transform
    dst_width = ref.width
    dst_height = ref.height
    profile_ref = ref.profile

print("Grid referencia (2014) ->", dst_width, "x", dst_height)


with rasterio.open(src_path) as src:
    src_data = src.read(1)
    src_nodata = src.nodata


    dst_data = np.full((dst_height, dst_width), src_nodata, dtype=src_data.dtype)

    reproject(
        source=src_data,
        destination=dst_data,
        src_transform=src.transform,
        src_crs=src.crs,
        dst_transform=dst_transform,
        dst_crs=dst_crs,
        src_nodata=src_nodata,
        dst_nodata=src_nodata,
        resampling=Resampling.bilinear
    )


    profile_new = profile_ref.copy()
    profile_new.update({
        "nodata": src_nodata
    })

    with rasterio.open(out_path, "w", **profile_new) as dst:
        dst.write(dst_data, 1)

print("LST_2024 alineado y guardado como:", out_path)
```

```python
import os
import rasterio
import numpy as np
from rasterio.warp import reproject, Resampling

base = "/content/drive/MyDrive/MAESTR√çA CIENCIA DE LOS DATOS Y ANAL√çTICA/TRABAJO DE GRADO/DATOS/DATOS_SATELITALES"
path_ndvi = os.path.join(base, "NDVI_2023.tif")
ref_path  = os.path.join(base, "LST_2014_Reproyectado.tif")  # raster de referencia
out_path  = os.path.join(base, "NDVI_2023_32618_30m.tif")


with rasterio.open(path_ndvi) as src:
    ndvi_raw = src.read(1).astype(float)
    ndvi_nodata = src.nodata
    crs_src = src.crs
    transform_src = src.transform
    print("NDVI_2023 CRS:", crs_src)
    print("NDVI_2023 Res:", src.res)


    vals = ndvi_raw[(ndvi_raw != ndvi_nodata) & (~np.isnan(ndvi_raw))]
    print("Rango NDVI_2023 (original):", vals.min(), vals.max())


    ndvi_norm = ndvi_raw.copy()


    if vals.max() > 2:
        print("Detectado NDVI escalado, normalizando dividiendo por 10000...")
        ndvi_norm = ndvi_raw / 10000.0


with rasterio.open(ref_path) as ref:
    dst_crs = ref.crs
    dst_transform = ref.transform
    dst_width = ref.width
    dst_height = ref.height
    profile_ref = ref.profile

print("Grid referencia (LST_2014) ->", dst_width, "x", dst_height, "| CRS:", dst_crs)


dst_data = np.full((dst_height, dst_width), ndvi_nodata if ndvi_nodata is not None else -9999, dtype=np.float32)

reproject(
    source=ndvi_norm,
    destination=dst_data,
    src_transform=transform_src,
    src_crs=crs_src,
    dst_transform=dst_transform,
    dst_crs=dst_crs,
    src_nodata=ndvi_nodata,
    dst_nodata=ndvi_nodata if ndvi_nodata is not None else -9999,
    resampling=Resampling.bilinear
)


profile_new = profile_ref.copy()
profile_new.update({
    "dtype": "float32",
    "nodata": ndvi_nodata if ndvi_nodata is not None else -9999
})

with rasterio.open(out_path, "w", **profile_new) as dst:
    dst.write(dst_data.astype(np.float32), 1)

print("NDVI continuo reproyectado guardado como:", out_path)

with rasterio.open(out_path) as src:
    arr = src.read(1).astype(float)
    nodata = src.nodata
    vals = arr[(arr != nodata) & (~np.isnan(arr))]
    print("Rango NDVI_2023_32618_30m:", vals.min(), vals.max())
```
